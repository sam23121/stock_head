# apply bag of words and random forest classification

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
cv = CountVectorizer()
X = cv.fit_transform(train_data)
```

```python
rfc = RandomForestClassifier(n_estimators=200, criterion='entropy')
```

```python
rfc.fit(X,train['Label'])
```

```python
test.head()
```

```python
test_data.columns = new_index
```

```python
for index in new_index:
    test_data[index] = test_data[index].str.lower()
```

```python
test_transform = []

for row in range(len(test_data)):
    test_transform.append(' '.join([str(i) for i in test_data.iloc[row,0:25]]))
test_transform[0]
```

```python
test_X = cv.transform(test_transform)
```

```python
predictions = rfc.predict(test_X)
```

```python
predictions
```

# model evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
cm = confusion_matrix(test['Label'], predictions)
cm
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title("confusion matrix")
plt.colorbar()
Labels = ['postive', 'negative']
tick_marks = np.arange(len(Labels))
plt.xticks(tick_marks, Labels, rotation = 45) 
plt.yticks(tick_marks, Labels)
plt.tight_layout()
plt.xlabel('True labels')
plt.ylabel('prediction label')

```

```python
accuracy = accuracy_score(test['Label'], predictions)
accuracy
```

```python
report = classification_report(test['Label'], predictions)
```

```python
print(report)
```
